```python
import numpy as np
import librosa
import matplotlib.pyplot as plt
def calculate_zero_crossing_rate(audio_file):
    signal,sample_rate=librosa.load(audio_file)
    zer=librosa.feature.zero_crossing_rate(signal)
    plt.figure(figsize=(12,6))
    plt.subplot(2,1,1)
    librosa.display.waveshow(signal,sr=sample_rate)
    plt.title("Waveform [43]")
    plt.subplot(2,1,1)
    plt.plot(zer[0])
    plt.title("Zero Crossing Rate [43]")
    plt.tight_layout()
    plt.show()
    return zer
if __name__=='__main__':
    audio_file='/content/example_WAV.wav'
    zer=calculate_zero_crossing_rate(audio_file)
    print(f"Zero Crossing Rate: \n'{zer}'")
```
